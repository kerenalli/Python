* Structured Query Language is known as SQL. 
* SQL is a computer language used to manage database data via SQL queries. 
* A table is a type of database object used by RDBMSs to store data. Multiple tables are frequently seen in a database. 
* A table name uniquely identifies each table, with linked data items arranged in rows and columns. Columns or fields, are vertical entities that hold all of the data related to a particular field in a table. 
* A table's rows are called records; they are the table's horizontal entities.

# **Basic SQL queries**
* INSERT
    * Syntax: INSERT INTO table_name (column1, column2, column3, ...) VALUES (value1, value2, value3, ...);
 * SELECT
    * Syntax: SELECT column1, column2, … FROM table_name;
        * SQL query: SELECT name FROM students;
    * Syntax: SELECT column1, column2, … FROM table_name WHERE condition; 
        * SELECT name, age FROM students WHERE age = 2000;
* AND, OR, NOT
    * Syntax: SELECT column1, column2, … FROM table_name WHERE condition1 AND condition2 AND condition3 AND condition4…..;
    * Syntax: SELECT column1, column2, … FROM table_name WHERE condition1 OR condition2 OR condition3 OR condition4…..;
    * Syntax: SELECT column1, column2, … FROM table_name WHERE NOT condition1;
* ORDER BY
    * Syntax: SELECT column1, column2, … FROM table_name ORDER BY column1, column2, ... ASC|DESC;
* GROUP BY
    * SELECT release_year, COUNT(*) AS number_of_movies FROM films GROUP BY release_year;
* UPDATE
    * Syntax: UPDATE table_name SET column1 = value1, column2 = value2, ...WHERE condition;
* DELETE
    * Syntax: DELETE FROM table_name WHERE condition;
* SQL Wildcard Characters 
    * LIKE: 
        * Syntax: SELECT column1, column2, … FROM table_name WHERE column1 LIKE patternToMatch;
     * Aggregate Functions:
         * COUNT, SUM, MIN, MAX,AVG

```python
import sqlite3 as sql

def create_table():
    conn = sql.connect("mydata.db")
    cur = conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS PERSON(SLN INT, NAME TEXT, EMAIL TEXT)")
    print("Table created successfully.")
    conn.close()
    
def insert_one_rec(S, N, E):
    conn = sql.connect("mydata.db")
    cur = conn.cursor()
    cur.execute("INSERT INTO PERSON(SLN, NAME, EMAIL) VALUES (?, ?, ?)", (S, N, E))
    print("Record inserted successfully.")
    conn.commit()  # Commit the transaction
    conn.close()

def insert_multi_rec(d_list):
    conn = sql.connect("mydata.db")
    cur = conn.cursor()
    cur.executemany("INSERT INTO PERSON(SLN, NAME, EMAIL) VALUES (?, ?, ?)", d_list)
    print("Records inserted successfully.")
    conn.commit()  # Commit the transaction
    conn.close()

def display_rec():
    conn = sql.connect("mydata.db")
    cur = conn.cursor()
    cur.execute("SELECT rowid, SLN, NAME, EMAIL FROM PERSON")
    res = cur.fetchall()
    print("Records fetched successfully.")
    for rec in res:
        print("Row ID:", rec[0])  # Print the rowid (first column)
        print("SLN:", rec[1])
        print("NAME:", rec[2])
        print("EMAIL:", rec[3])
        print()  # Add a blank line for better readability
    conn.close()
    
def update_rec():
    conn = sql.connect("mydata.db")
    cur = conn.cursor()
    c = "UPDATE PERSON SET NAME = ? WHERE SLN = ?"  # Corrected query string
    cur.execute(c, ("Bhuvan", 231))
    conn.commit()
    print("Record updated successfully.")
    conn.close()
    
```

```python
# Create table
create_table()

# Insert single record
insert_one_rec(1234, "Sudarshana", "ksudarsh@gitam.in")

# Insert multiple records
d_list = [
    (123, "sudarshana", "kerenalli@gmail.com"),
    (231, "bhuvan", "bhuvan@gmail.com"),
    (456, "Jagan", "jagank@hotmail.com")
]
insert_multi_rec(d_list)

# Display records
display_rec()
update_rec()
```

# **Advanced SQL Commands**
* Different types of SQL Joins
    * Inner Join-Common records from both the tables
    * Left (Outer) Join: records from left table and common from right one
    * Right (Outer) Join: records from right table and common from left one
    * Full (Outer) Join: Returns all records when there is a match in either left or right table
