# Trigonometric Functions Package

```python
!mkdir trig_pack
!touch trig_pack/__init__.py
!touch trig_pack/trig_funs.py
```

```python
txt = '''
import math

def custom_sin(x):
    return math.sin(x)

def custom_cos(x):
    return math.cos(x)

def custom_tan(x):
    return math.tan(x)
'''
with open('/kaggle/working/trig_pack/trig_funs.py','w') as f:
    f.write(txt)
    f.close()
    
```

```python
from trig_pack import trig_funs as tfs
import math
x = int(input("Enter the angle! "))
print(f"Sin({x}) is: {tfs.custom_sin(x)}")
print(f"Cos({x}) is: {tfs.custom_cos(x)}")
print(f"Tan({x}) is: {tfs.custom_tan(x)}")
print(math.sin(x))
```
