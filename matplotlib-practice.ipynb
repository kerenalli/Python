* Data Visualization:
    * The process of conveying the information with the help of plots and graphics.
    * The goal of data visualization is to communicate information in a clear and efficient manner.
    * Matplotlib is the basic data visualization tool of Python programming language. 
    * The project Matplotlib was started by **John Hunter** in **2002**.
 * Matplotlib Object Hierarchy
    * It is an excellent 2D and 3D graphics library for generating scientific figures.
    * In Matplotlib, a plot is a hierarchy of nested Python objects.
        * A Figure object is the outermost container for a Matplotlib plot.
            * The Figure object can be imagined as a box-like container containing one or more Axes.
        * The Figure object contain multiple Axes objects.
            * The Axes object contain smaller objects such as tick marks, lines, legends, title and text-boxes.
            * Plots in Matplotlib reside within a Figure object. 
        * Matplotlib.pyplot is a collection of commands and functions that make Matplotlib behave like MATLAB (for plotting). 
        * The MATLAB-style tools are contained in the pyplot (plt) interface.
* Parts of a Plot
    * There are different parts of a plot. 
        * These are title, legend, grid, axis and labels etc.

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.plot(x,y,
         marker = 'p',
         markersize = 20,
         markerfacecolor = "#0ABCDE",
         markeredgecolor = "#0FFDFA",
         linestyle = '-.',
         linewidth = 5,
         color = "orange"
        )
plt.title("Product Chart")
plt.xlabel("Number")
plt.ylabel("Result")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(0,10,15)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y, label = "Sin(x)")
plt.plot(x,y2,label = "cos(x)")
plt.xlabel("X")
plt.ylabel("Sine x")
plt.legend(title ="Waves")
plt.title("Sin and Cos Chart")
plt.show()
plt.savefig("sincoschart.pdf")
```

```python
x = np.linspace(0,10,20)
y = np.sin(x)

fig_1 = plt.figure(figsize = (5,4))
axes_1 = fig_1.add_axes([0.4,0.6,0.4,0.5])
axes_1.set_xlabel("x")
axes_1.set_ylabel("sin(x)")
axes_1.set_title("Sine Chart")
axes_1.plot(x,y)
plt.show()
plt.savefig("sine.pdf")
```

```python
#Pie Chart
Expenses = np.array([40, 30,10,20])
Labels = np.array(["Rent","Food","Transit","Misc"])
Colors = np.array(["r","g","y","m"])
Explode = np.array([0, 0 , 0,.20])
plt.pie(Expenses,labels = Labels,explode = Explode,colors = Colors,startangle = 90)
plt.show()
```

# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1,5,25)
y1 = x**2
y2 = x**3
y3 = x**4
plt.subplot(1,3,1)
plt.plot(x,y1,label = "x^2")
plt.xlabel("x")
plt.ylabel("y^2")
plt.title("X sqaure chart")
plt.subplot(1,3,2)
plt.plot(x,y2,label = "x^3")
plt.xlabel("x")
plt.ylabel("y^3")
plt.title("X cube chart")
plt.subplot(1,3,3)
plt.plot(x,y3,label = "x^4")
plt.xlabel("x")
plt.ylabel("y^4")
plt.title("X^4 chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1,5,25)
y1 = x**2
y2 = x**3
y3 = x**4
plt.subplot(2,2,1)
plt.plot(x,y1,label = "x^2")
plt.xlabel("x")
plt.ylabel("y^2")
plt.title("X sqaure chart")
plt.subplot(2,2,2)
plt.plot(x,y2,label = "x^3")
plt.xlabel("x")
plt.ylabel("y^3")
plt.title("X cube chart")

plt.subplot(2,2,3)
plt.plot(x,y3,label = "x^4")
plt.xlabel("x")
plt.ylabel("y^4")
plt.title("X^4 chart")
plt.subplot(2,2,4)
plt.plot(x,x,label="x")
plt.tight_layout()
plt.show()
```

**1. Simple Functional Plot**

```python
x_series = [1,2,3,4,5,6]
y_series = [1,4,9,16,25,36]
plt.plot(x_series, y_series,"bd--")
#Add the title to the graph
plt.title("Number vs Square")
plt.plot(x_series, x_series, "r*-")
#Add the x label
plt.xlabel("Number")
#Add the y label
plt.ylabel("Square")
plt.show()
```

**2. Multiple Plots**
* Subplots

```python
plt.subplot(1,2,1)
plt.plot(x_series,x_series,"b*-")
plt.xlabel("The number")
plt.ylabel("The Number")
plt.title("Number vs Number")
plt.subplot(1,2,2)
plt.plot(x_series, y_series, marker = "*",ms = 12, mfc = "r")
plt.xlabel("The number")
plt.ylabel("The Square")
plt.title("Number Vs Square")
#plt.savefig("img.pdf")
plt.show()
```

# Figure and Axes

```python
fig_1 = plt.figure(figsize = (5,4))
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel("Number")
axes_1.set_ylabel("Square")
axes_1.set_title("Number vs square")
axes_1.plot(x_series,y_series,label = "x vs x**2",marker = 'o',linewidth = 5, ms = 12,mfc = 'r')
axes_1.plot(y_series,x_series,label = "x**2 vs x",marker = '*',linewidth = 4, ms = 12, mfc = 'g')
axes_1.legend(loc=0)
plt.show()
```

**Adding multiple axes**

# Creating Pie Charts

```python
results = np.array([20,40,10,10,20 ])
sections = np.array(['AIML','CSE','IoT','CSBS','DS'])
ex = np.array([0,0.2,0,0,0])
col = np.array(['r','k','b','m','g'])
plt.pie(results,
       labels = sections,
       startangle = 90,
       explode = ex,
       shadow = True,
       colors = col,
       )
plt.legend(title = "Sections",loc =0)
plt.show()

```
